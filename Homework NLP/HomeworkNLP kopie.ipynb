### Stap 1 : data inladen

De data bestaat uit een .csv file die rechtstreeks ingeladen kan worden. De belangrijkste kolommen zijn uiteraard de _Review_text_ en de _Rating_ kolom. De ratings gaan van een waarde 1 tot 5, waarbij 5 uitstekend is (heel positieve review) en 1 een slechte review. </br>Modeleer hieruit een classificatieprobleem (je hoeft zeker geen 5 klassen te behouden, herwerk bvb naar een binaire classificatie of eventueel naar een probleem met 3 klassen (positief neutraal, negatief).

```python
#Importing data
import pandas as pd
#Read csv data and put in dataframe
df = pd.read_csv("../data/DisneylandReviews.csv",encoding='latin-1')
#Print full dataframe
#Print(df)
#Print only head to see structure
df.head()
df
```

```python
#NLTK installation
import nltk
print(nltk.__version__)
```

```python
#Tokenization
from nltk.tokenize import word_tokenize
#nltk.download('punkt')
#Add new column with tokenized reviews
#src: https://stackoverflow.com/questions/33098040/how-to-use-word-tokenize-in-data-frame
df_clean['tokenized_sents'] = df_clean.apply(lambda row: word_tokenize(row['Review_Text']), axis=1)
df_clean
```

```python
#Punctual removal + lower casing
#src: https://stackoverflow.com/questions/15547409/how-to-get-rid-of-punctuation-using-nltk-tokenizer
from nltk.tokenize import RegexpTokenizer

def remove_punctuations(text):
    tokenizer = RegexpTokenizer(r'\w+')
    tokenized = tokenizer.tokenize(text)
    return [w.lower() for w in tokenized if w.isalpha()]
    
df_clean["tokenized"] = df_clean.apply(lambda row: remove_punctuations(row['Review_Text']), axis=1)
#We zien nu dat tokenizen en punctuations samen gebeurt dus was de vorige stap overbodig.
df_clean.drop(['tokenized_sents'], axis=1, inplace=True)
df_clean
```

```python
#Stopword removal
#nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
#Deleting stopwords from our tokenized sentences
#src: https://stackoverflow.com/questions/29523254/python-remove-stop-words-from-pandas-dataframe
def stopwords(text):
    return [word for word in text if not word in stop_words]
    
df_clean['tokenized_without_stopwords'] = df_clean.apply(lambda x: stopwords(x['tokenized']), axis=1)

print('Tokenized text: ', df_clean['tokenized'].head())
print()
print('With stopwords removed: ', df_clean['tokenized_without_stopwords'].head())
```

```python
#Overview
df_clean
```

```python
#Lemmatizing
#src:https://www.datacamp.com/community/tutorials/text-analytics-beginners-nltk
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')

def lemmanize(text):
    lem = WordNetLemmatizer()
    return[lem.lemmatize(word) for word in text]
    
df_clean['Lemmanized'] = df_clean.apply(lambda x: lemmanize(x['tokenized_without_stopwords']), axis=1)
df_clean
```

```python
#This is what lemmanizing does
lem = WordNetLemmatizer()
word = "flying"
print("Lemmatized Word:",lem.lemmatize(word,"v"))
```

```python
#Stemming
#src: https://www.datacamp.com/community/tutorials/text-analytics-beginners-nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

def stemming(text):
    ps = PorterStemmer()
    return[ps.stem(word) for word in text]

df_clean['Stemmed'] = df_clean.apply(lambda x: stemming(x['tokenized_without_stopwords']), axis=1)

df_clean.head()
```

```python
#This is what stemming does
ps = PorterStemmer()
word = "flying"
print("Stemmed Word:",ps.stem(word))
```

```python
#Cleaning dataframe for models
df_clean.drop(['Review_Text','tokenized','tokenized_without_stopwords','Stemmed'], axis=1, inplace=True)
df_clean
```

```python
df_clean['Cleaned_Reviews'] = df_clean['Lemmanized']
df_clean.drop(['Lemmanized'], axis=1, inplace=True)
df_clean
```

```python
#src: https://www.geeksforgeeks.org/python-program-to-convert-a-list-to-string/
def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # return string  
    return (str1.join(s))

df_clean['Cleaned_Reviews_String'] = df_clean.apply(lambda x: listToString(x['Cleaned_Reviews']), axis=1)

df_clean
```

```python
#Make sentiment int for classification later
pos = [5]
neg = [1, 2]
neu = [3, 4]

def sentiment(rating):
    if rating in pos:
        return 2
    elif rating in neg:
        return 0
    else:
        return 1  
df_clean['Sentiment_INT'] = df_clean['Rating'].apply(sentiment)
df_clean.head()
```

```python
#VADER
#Since VADER is pretrained, you can get results more quickly than with many other analyzers. However, VADER is best suited for language used in social media, like short sentences with some slang and abbreviations. Itâ€™s less accurate when rating longer, structured sentences, but itâ€™s often a good launching point.
#We gebruiken de naam df_vader voor dit model.
df_vader = df_clean
df_vader
```

```python
#VADER
#src:https://realpython.com/python-nltk-sentiment-analysis/#using-nltks-pre-trained-sentiment-analyzer
from nltk.sentiment import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

df_vader['result'] = df_vader['Cleaned_Reviews_String'].apply(lambda x: sia.polarity_scores(x))
df_vader
```

```python
#Bag of words
#Rename df_vader to df_words for bag of words
df_words = df_vader
df_words
```

```python
#Count vectorizer
#src: https://www.mygreatlearning.com/blog/bag-of-words/
#src: https://vitalflux.com/text-classification-bag-of-words-model-python-sklearn/
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

X = df_words['Cleaned_Reviews_String'].values.tolist()
y = df_words['Sentiment_INT'].values.tolist()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

vectorizer = CountVectorizer()
count_train = vectorizer.fit_transform(X_train)
count_test = vectorizer.transform(X_test)
```

```python
#TFIDF vectorizer
#src: https://machinelearningmastery.com/prepare-text-data-machine-learning-scikit-learn/

from sklearn.feature_extraction.text import TfidfVectorizer

tfvector = TfidfVectorizer()
tfvector_train = tfvector.fit_transform(X_train)
tfvector_test = tfvector.transform(X_test)
```

```python
print(tfvector_train.shape, tfvector_train.shape)
```

```python
#Use classification model to train and test data using the count vectorizer data

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

lr = LogisticRegression(random_state=1, solver='lbfgs', multi_class='ovr')

lr.fit(count_train, y_train)

y_predict = lr.predict(count_test)

print("LogisticRegression Accuracy %.3f" %metrics.accuracy_score(y_test, y_predict))
```

```python
#Use classification model to train and test data using the tgif vectorizer data

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

lr = LogisticRegression(random_state=1, solver='lbfgs', multi_class='ovr')

lr.fit(tfvector_train, y_train)

y_predict = lr.predict(tfvector_test)

print("LogisticRegression Accuracy %.3f" %metrics.accuracy_score(y_test, y_predict))
```

```python
#Deep learning/RNN
#src: https://stackabuse.com/python-for-nlp-movie-sentiment-analysis-using-deep-learning-in-keras/
df_rnn = df_words
df_rnn.head()
```

```python
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten
from keras.layers import GlobalMaxPooling1D
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
```

```python
#X = df_words['Cleaned_Reviews_String'].values.tolist()
#y = df_words['Sentiment_INT'].values.tolist()
X = np.array(df_words['Cleaned_Reviews_String'].values)
y = np.array(df_words['Sentiment_INT'].values)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)

X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)

vocab_size = len(tokenizer.word_index) + 1

maxlen = 100

X_train = pad_sequences(X_train, padding='post', maxlen=maxlen)
X_test = pad_sequences(X_test, padding='post', maxlen=maxlen)
```

```python
from numpy import array
from numpy import asarray
from numpy import zeros

#embeddings_dictionary = dict()
#glove_file = open('E:/Datasets/Word Embeddings/glove.6B.100d.txt', encoding="utf8")

#for line in glove_file:
#    records = line.split()
#    word = records[0]
#    vector_dimensions = asarray(records[1:], dtype='float32')
#    embeddings_dictionary [word] = vector_dimensions
#glove_file.close()
embedding_matrix = zeros((vocab_size, 100))
#for word, index in tokenizer.word_index.items():
#    embedding_vector = embeddings_dictionary.get(word)
#    if embedding_vector is not None:
#        embedding_matrix[index] = embedding_vector
```

```python
model = Sequential()
embedding_layer = Embedding(vocab_size, 100, weights=[embedding_matrix], input_length=maxlen , trainable=False)
model.add(embedding_layer)

model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

print(model.summary())
```

```python
history = model.fit(X_train, y_train, batch_size=128, epochs=6, verbose=1)
```

```python
score = model.evaluate(X_test, y_test, verbose=1)
```

```python
print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
